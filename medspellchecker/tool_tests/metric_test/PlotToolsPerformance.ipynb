```python
def plot_global_performance(x,y,point_labels, x_label, y_label, title, ncol, position):
    plt.figure(figsize=(10,10))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    for idx in range(len(x)):
        if "MedSpellChecker" in point_labels[idx]:
            plt.scatter(x[idx], math.log(y[idx]), marker="$M$")
        elif "Yandex" in point_labels[idx]:
            plt.scatter(x[idx], math.log(y[idx]), marker="$Y$")
        else:
            plt.scatter(x[idx], math.log(y[idx]))

    for i, name in enumerate(point_labels):
        plt.annotate(i+1, (x[i]+1, y[i]+1))
    plt.legend(labels=point_labels, loc=position, ncol=ncol, bbox_to_anchor=(1.42, 0.5))
    plt.show()
```

```python
plot_global_performance(overall_precisions, words_per_second, tool_names, "Overall precision", "Logarithm of words per second", "", 1, "center right")
```

```python
def plot_precisions(x,y,point_labels, x_label, y_label, title, ncol, position):
    plt.figure(figsize=(10,10))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    for idx in range(len(x)):
        if "MedSpellChecker" in point_labels[idx]:
            plt.scatter(x[idx], y[idx], marker="$M$")
        elif "Yandex" in point_labels[idx]:
            plt.scatter(x[idx], y[idx], marker="$Y$")
        else:
            plt.scatter(x[idx], y[idx])

    for i, name in enumerate(point_labels):
        plt.annotate(i+1, (x[i]+1, y[i]+1))
    plt.legend(labels=point_labels, loc=position, ncol=ncol, bbox_to_anchor=(1.42, 0.5))
    plt.show()
```

```python
plot_precisions(error_precision, lexical_precision, tool_names, "Error precision", "Lexical precision", "", 1, 'center right')
```
