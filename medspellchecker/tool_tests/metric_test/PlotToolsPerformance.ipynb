```python
from matplotlib import pyplot as plt
import math
```

```python
plt.rcParams.update({'font.size': 14})
```

```python
# OVERALL - WPS - ERROR - LEXICAL - NAME - SYMBOL
tools = [
    (0.831, 357.3, 0.731, 0.93,"Aspell-python", '$A$'),
    (0.713, 11.8, 0.706, 0.719, "PyHunspell", '$H$'),
    (0.72, 24.3, 0.721, 0.719, "PyEnchant", '$E$'),
    (0.835, 43.6, 0.727, 0.942, "LanguageTool-python", '$L$'),
    (0.586, 6.7, 0.304, 0.868, "PySpellChecker", '$C$'),
    (0.642, 26060.2, 0.37, 0.913, "SymSpellPy", '$S$'),
    (0.643, 1604.2, 0.483, 0.804, "SymSpellPy (compound)", '$S$'),
    (0.638, 4322.3, 0.307, 0.969, "JumSpell", '$J$'),
    (0.904, 1308.8, 0.82, 0.987, "Yandex Speller", '$Y$'),
    (0.888, 8.9, 0.792, 0.984, "MedSpellChecker\n(MedRuRoberta/CPU)", '$M$'),
    (0.878, 45.3, 0.765, 0.99, "MedSpellChecker\n(MedDistilBert/CPU)", '$M$'),
    (0.865, 127.3, 0.742, 0.987, "MedSpellChecker\n(MedRuBertTiny2/CPU)", '$M$'),
    (0.863, 9.1, 0.738, 0.987,"MedSpellChecker\n(RuBioRoBerta/CPU)", '$M$'),
    (0.852, 30.7, 0.715, 0.988, "MedSpellChecker\n(RuBioBerta/CPU)", '$M$'),
]

overall_precisions = [item[0] for item in tools]
words_per_second = [item[1] for item in tools]
error_precision = [item[2] for item in tools]
lexical_precision = [item[3] for item in tools]
tool_names = [item[4] for item in tools]
tool_symbols = [item[5] for item in tools]
names_to_symbol = {item[4]:item[5] for item in tools}
```

```python
def plot_metrics(x, y, x_label, y_label, title, ncol, position):
    plt.figure(figsize=(9,9))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    for idx in range(len(x)):
        plt.scatter(x[idx], math.log(y[idx]), marker=names_to_symbol[tool_names[idx]])

    for i, name in enumerate(tool_names):
        plt.annotate(i+1, (x[i]+1, y[i]+1))
    plt.legend(labels=tool_names, loc=position, ncol=ncol, bbox_to_anchor=(1.47, 0.5))
    plt.show()
```

```python
plot_metrics(overall_precisions, words_per_second, "Overall precision", "Log of words per second", "", 1, "center right")
```

```python
def plot_precisions(x, y, x_label, y_label, title, ncol, position):
    plt.figure(figsize=(9,9))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    for idx in range(len(x)):
        plt.scatter(x[idx], y[idx], marker=names_to_symbol[tool_names[idx]])

    for i, name in enumerate(tool_names):
        plt.annotate(i+1, (x[i]+1, y[i]+1))
    plt.legend(labels=tool_names, loc=position, ncol=ncol, bbox_to_anchor=(1.47, 0.5))
    plt.show()
```

```python
plot_precisions(error_precision, lexical_precision, "Error precision", "Lexical precision", "", 1, 'center right')
```
