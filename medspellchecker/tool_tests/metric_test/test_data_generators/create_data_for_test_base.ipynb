```python
import pandas as pd
import pickle
import numpy as np
```

```python
random_state = 111
```

```python
df_all_anamnesis = pd.read_csv("../../../../data/anamnesis/processed/all_anamnesis.csv", names=["text"])
df_all_anamnesis
```

```python
all_processed_anam = df_all_anamnesis["text"].apply(lambda x: x.split())
```

```python
import itertools

BATCH_SIZE = 10
batches = []

for anamnes in all_processed_anam:
    for i in range(len(anamnes) - BATCH_SIZE):
        batches.append(list(itertools.islice(anamnes, i, i+BATCH_SIZE)))
```

```python
batches[:10]
```

```python
import random

random.seed(random_state)

NUMBER_BATCHES = 400

picked_batches = random.sample(batches, 400)
picked_batches
```

```python
len(picked_batches)
```

```python
picked_batches = [' '.join(batch) for batch in picked_batches]
picked_batches
```

```python
dataframe = pd.DataFrame({"original" : picked_batches, "answer": picked_batches, "pos_incorrect_word": [0]*NUMBER_BATCHES})
dataframe
```

```python
dataframe.to_csv("../../../../data/test/with_context/data_for_test_with_context_new.csv", index=False)
```

```python
print(f"Row ID --- Original --- Answer")
counter = 0
for i in range(len(dataframe)):
    original = dataframe.original[i].split()
    answer = dataframe.answer[i].split()
    pos_incorrect_word = int(dataframe.pos_incorrect_word[i])
    if original[pos_incorrect_word] != answer[pos_incorrect_word]:
        counter += 1
        print(f"{i} --- {original[pos_incorrect_word]} --- {answer[pos_incorrect_word]}")
    else:
        print(f"!!! {i} --- {original[pos_incorrect_word]} --- {answer[pos_incorrect_word]} !!!")
print(f"Total record {counter}")
```
