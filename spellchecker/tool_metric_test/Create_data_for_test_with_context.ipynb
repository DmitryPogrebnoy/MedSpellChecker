```python
import pandas as pd
import pickle
import numpy as np
```

```python
random_state = 111
```

```python
all_anam = pickle.load(open('../../data/all_anam.pickle', 'rb'))
all_anam
```

```python
from pre_post_processor import PreProcessor

pre_processor = PreProcessor()

all_processed_anam = [pre_processor.lemmatize(word) for word in all_anam if pre_processor._is_valid_token(word)]
all_processed_anam
```

```python
len(all_processed_anam)
```

```python
import itertools

BATCH_SIZE = 10
batched_all_processed_anam = [[batch for batch in itertools.islice(all_processed_anam, i, i+BATCH_SIZE)]
          for i in range(len(all_processed_anam) - BATCH_SIZE)]
batched_all_processed_anam
```

```python
import random

random.seed(random_state)

NUMBER_BATCHES = 400

picked_batches = random.sample(batched_all_processed_anam, 400)
picked_batches
```

```python
len(picked_batches)
```

```python
picked_batches = [ ' '.join(batch) for batch in picked_batches]
picked_batches
```

```python
dataframe = pd.DataFrame({"original" : picked_batches, "answer": picked_batches, "pos_incorrect_word": [0]*NUMBER_BATCHES})
dataframe
```

```python
dataframe.to_csv("../../data/data_for_test_with_context_new.csv")
```

Let's validate data

```python
dataframe = pd.read_csv("../../data/data_for_test_with_context.csv", index_col=0)
dataframe = dataframe.reset_index(drop=True)
dataframe
```

```python
print(f"Row ID --- Original --- Answer")
counter = 0
for i in range(len(dataframe)):
    original = dataframe.original[i].split()
    answer = dataframe.answer[i].split()
    pos_incorrect_word = dataframe.pos_incorrect_word[i]
    if original[pos_incorrect_word] != answer[pos_incorrect_word]:
        counter += 1
        print(f"{i} --- {original[pos_incorrect_word]} --- {answer[pos_incorrect_word]}")
    else:
        print(f"!!! {i} --- {original[pos_incorrect_word]} --- {answer[pos_incorrect_word]} !!!")
print(f"Total record {counter}")
```

Check that no other different words in batch except one incorrect

```python
print(f"Row ID --- Original word --- Answer word")
counter = 0
for i in range(len(dataframe)):
    original = dataframe.original[i].split()
    answer = dataframe.answer[i].split()
    pos_incorrect_word = dataframe.pos_incorrect_word[i]
    for j in range(BATCH_SIZE):
        if j != pos_incorrect_word and original[j] != answer[j]:
            print(f"{i} --- {original[j]} --- {answer[j]}")
print(f"If output is clear that it's ok")
```

Check that words exist in global dict

```python
with open("../../data/dictionaries/processed/processed_lemmatized_all_dict.txt") as file:
    global_dict = file.readlines()
    global_dict = set([word[:-1] for word in global_dict])
print(f"Row ID --- Unknown word")
counter = 0
for i in range(len(dataframe)):
    original = dataframe.original[i].split()
    pos_incorrect_word = dataframe.pos_incorrect_word[i]
    for j in range(BATCH_SIZE):
        if j != pos_incorrect_word and original[j] not in global_dict:
            print(f"{i} --- {original[j]}")
print(f"If output is clear that it's ok")
```
