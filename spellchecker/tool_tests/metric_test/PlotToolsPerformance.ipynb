```python
def plot(x,y,point_labels, x_label, y_label, title, ncol, position):
    plt.figure(figsize=(10,10))
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    for idx in range(len(x)):
        if "MedSpellChecker" in point_labels[idx]:
            plt.scatter(x[idx], y[idx], marker="x")
        elif "Yandex" in point_labels[idx]:
            plt.scatter(x[idx], y[idx], marker="$Y$")
        else:
            plt.scatter(x[idx], y[idx])

    for i, name in enumerate(point_labels):
        plt.annotate(i+1, (x[i]+1, y[i]+1))
    plt.legend(labels=point_labels, loc=position, ncol=ncol)
    plt.show()
```
